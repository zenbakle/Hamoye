```python
# importing packages 
import pandas as pd
import numpy as np
```

```python
# importing dataset and viewing first five column 
fuel_ferc = pd.read_csv("fuel_ferc1.csv")
fuel_ferc.head()
```

```python
# Average cost of fuel per mmbtu by type of fuel
fuel_ferc.groupby('fuel_type_code_pudl')["fuel_cost_per_mmbtu"].mean()
```

```python
# summary statistics for dataset 
fuel_ferc.describe()
```

```python
# skewness
fuel_ferc.skew(axis =0, skipna = True)
```

```python
# kurtosis
fuel_ferc.kurtosis(axis=0, skipna =True)
```

```python
# information on dataset
fuel_ferc.info()
```

```python
# view all the rows that have missing values based on fuel_unit
missing = fuel_ferc[fuel_ferc["fuel_unit"].isnull()].count()
```

```python
# percentage of missing values 
lenght = len(fuel_ferc)
per = missing/lenght
per
```

```python
#correlation of all columns
fuel_ferc.corr()
```
